### Задание 2.

```python

import pandas as pd
import requests

params = {
    "q": "language:JavaScript",
    "sort": "stars",
    "per_page": 20
}

r = requests.get(url+"search/repositories", params=params).json()
repos = r["items"]
# Проверяем, что вернулся корректный JSON
if "items" in r:
    repos = r["items"]
    df = pd.DataFrame(repos)[["name", "stargazers_count"]].sort_values(
        "stargazers_count", ascending=False
    )
    
    # Сбросить индекс и начать с 1
    df = df.reset_index(drop=True)
    df.index += 1
    
    print(df.head(10))
else:
    print("Ошибка запроса:", r)
```
